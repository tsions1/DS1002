```python
!pip install english_words
from english_words import get_english_words_set
words = get_english_words_set(['web2'], lower=True)
```

```python
#convert words into a list
wordlist = list(words)

#print the first ten words
print(wordlist[:10])

#make list called alpha containing the words in alphabetical order
alpha_list = sorted(wordlist)
print(alpha_list[:10])

#counting the words that start with a
count = sum(word.startswith('a') for word in alpha_list)
print(count)

#counting the words that start with q
count1 = sum(word.startswith('q') for word in alpha_list)
print (count1)

#fraction of words without e
without_e = sum('e' not in word for word in alpha_list)
fraction_without_e = without_e / len(alpha_list)
print(fraction_without_e)

#convert into NumPy arrays
import numpy as np
wordlist_np = np.array(wordlist)
alpha_list_np = np.array(alpha_list)

alphabet = [chr(i) for i in range(ord('a'),ord('z')+1)]
print(alphabet)

#word count for each word
word_count = {}
for letter in alphabet:
  word_count[letter] = sum(word.startswith(letter) for word in alpha_list)
print(word_count)

#bar graph
import matplotlib.pyplot as plt
plt.bar(word_count.keys(), word_count.values())
plt.xlabel('Letter')
plt.ylabel('Number of words')
plt.title('Number of words starting with each letter')
plt.show()

#six longest words
longest_words = sorted(wordlist, key=len, reverse=True)[:6]
print(longest_words)

#histogram
word_length = [len(word) for word in wordlist]
plt.hist(word_length, bins=range(min(word_length), max(word_length) + 1))
plt.xlabel('word length')
plt.ylabel('frequency')
plt.title('Number of words of each length')
plt.show()
```

```python
#create list containing all the pairs
pairs_list = []

for letter1 in alphabet:
  for letter2 in alphabet:
    pairs_list.append(letter1 + letter2)

pair_freq = {}


for pair in pairs_list:
  pair_freq[pair] = 0

# Set the frequency for all pairs to 0
for word in alpha_list:
  for pair in pairs_list:
    if pair in word:
      pair_freq[pair] += 1
```
